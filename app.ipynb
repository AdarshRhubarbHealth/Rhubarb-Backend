```python
# Function to check if an ingredient is present in the list
def check_ingredient(ingredient_list, flag_name, flag_value):
    if flag_value:
        for ingredient in ingredient_list:
            if ingredient.lower() in [flag_name.lower(), flag_name.lower() + 's']:
                return True
    return False
```

```python
# Get user input
user_full_name = input("Enter your full name: ")

allergies = input("Do you have any allergies? (yes/no): ").lower() == 'yes'
preservatives_flag = input("Are you concerned about preservatives? (yes/no): ").lower() == 'yes'
sugar_content_flag = input("Are you concerned about sugar content? (yes/no): ").lower() == 'yes'
inverted_sugar = input("Are you concerned about inverted sugar? (yes/no): ").lower() == 'yes'
gmo_flag = input("Are you concerned about GMOs? (yes/no): ").lower() == 'yes'
usda_organic_seal = input("Are you concerned about USDA Organic Seal? (yes/no): ").lower() == 'yes'
gmo = input("Are you concerned about GMO? (yes/no): ").lower() == 'yes'
xanthan_gum = input("Are you concerned about xanthan gum? (yes/no): ").lower() == 'yes'


ingredient_list = input("Enter a list of ingredients (separated by commas): ").lower().split(',')
# Check if any flagged ingredient is present
allergen_present = check_ingredient(ingredient_list, 'allergies', allergies)
preservative_present = check_ingredient(ingredient_list, 'preservative', preservatives_flag)
sugar_present = check_ingredient(ingredient_list, 'sugar', sugar_content_flag)
inverted_sugar_present = check_ingredient(ingredient_list, 'inverted sugar', inverted_sugar)
gmo_present = check_ingredient(ingredient_list, 'gmo', gmo_flag or gmo)
usda_organic_seal_present = check_ingredient(ingredient_list, 'usda organic', usda_organic_seal)
xanthan_gum_present = check_ingredient(ingredient_list, 'xanthan gum', xanthan_gum)

# Display the signal based on the presence of flagged ingredients
if allergen_present or preservative_present or sugar_present or inverted_sugar_present or gmo_present or usda_organic_seal_present or xanthan_gum_present:
    print("Red signal: One or more flagged ingredients found in the list.")
else:
    print("Green signal: No flagged ingredients found in the list.")


```
